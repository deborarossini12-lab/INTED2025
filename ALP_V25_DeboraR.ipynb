Peça a quantidade de horas trabalhadas e o valor por hora. Se trabalhou até 40 horas, calcule o salário normalmente (horas × valor por hora). Se trabalhou mais de 40 horas, as horas extras devem ter um adicional de 50% (horas extras × valor por hora × 1.5). Crie um algoritmo em Python que realize esse cálculo corretamente.

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor por hora: "))

if horas_trabalhadas <= 40:
  salario = horas_trabalhadas * valor_por_hora
else:
  horas_extras = horas_trabalhadas - 40
  salario = (40 * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)

print("O salário é:", salario)
```

Crie um algoritmo que faça uma calculadora, que repita até que o usuário digite 0 para sair:

Deve mostrar o seguinte menu na tela:

Calculadora V25

Digite:

1 - Multiplos de 5

2 - Números Pares

3 - Números Primos

4 - Conversor de temperatura

0 - Sair

------------------------------------------------------

Cada vez que o usuário digitar um número deve entrar na função e pedir os dados ao usuário e efetuar o calculo.

mostrar a resposta e solicitar novamente o que ele quer fazer.

```python
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

while True:
    print("\nCalculadora V25")
    print("Digite:")
    print("1 - Multiplos de 5")
    print("2 - Números Pares")
    print("3 - Números Primos")
    print("4 - Conversor de temperatura")
    print("0 - Sair")
    print("------------------------------------------------------")

    choice = input("Escolha uma opção: ")

    if choice == '0':
        print("Saindo da calculadora. Até mais!")
        break
    elif choice == '1':
        try:
            num = int(input("Digite um número para verificar múltiplos de 5: "))
            if num % 5 == 0:
                print(f"O número {num} é múltiplo de 5.")
            else:
                print(f"O número {num} NÃO é múltiplo de 5.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")
    elif choice == '2':
        try:
            num = int(input("Digite um número para verificar se é par: "))
            if num % 2 == 0:
                print(f"O número {num} é par.")
            else:
                print(f"O número {num} é ímpar.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")
    elif choice == '3':
        try:
            num = int(input("Digite um número para verificar se é primo: "))
            if is_prime(num):
                print(f"O número {num} É primo.")
            else:
                print(f"O número {num} NÃO é primo.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um número inteiro.")
    elif choice == '4':
        try:
            temp = float(input("Digite a temperatura em Celsius para converter para Fahrenheit: "))
            fahrenheit = (temp * 9/5) + 32
            print(f"{temp}°C é igual a {fahrenheit}°F.")
        except ValueError:
            print("Entrada inválida. Por favor, digite um valor numérico para a temperatura.")
    else:
        print("Opção inválida. Por favor, escolha uma opção válida do menu.")
```
